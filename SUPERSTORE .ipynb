```python


import pandas as pd

# Load dataset
df = pd.read_excel("/Users/rajanandinialla/Documents/data sets/Sample - Superstore.xls")

# View the first few rows
print(df.head())
```

```python
# Check for missing values
print(df.isnull().sum())

# Data types
print(df.dtypes)

# Descriptive statistics
print(df.describe())
```

```python
# Drop rows with missing values (if necessary)
df = df.dropna()

# Fill missing values with a placeholder (if appropriate)
df['Row ID'].fillna('Unknown', inplace=True)
df['Order ID'].fillna('Unknown', inplace=True)
df['Order Date'].fillna('Unknown', inplace=True)
df['Ship Date'].fillna('Unknown', inplace=True)
df['Ship Mode'].fillna('Unknown', inplace=True)
df['Customer ID'].fillna('Unknown', inplace=True)
df['Customer Name'].fillna('Unknown', inplace=True)
df['Segment'].fillna('Unknown', inplace=True)
df['Country/Region'].fillna('Unknown', inplace=True)
df['City'].fillna('Unknown', inplace=True)
df['State/Province'].fillna('Unknown', inplace=True)
df['Postal Code'].fillna('Unknown', inplace=True)
df['Region'].fillna('Unknown', inplace=True)
df['Product ID'].fillna('Unknown', inplace=True)
df['Category'].fillna('Unknown', inplace=True)
df['Sub-Category'].fillna('Unknown', inplace=True)
df['Product Name'].fillna('Unknown', inplace=True)
df['Quantity'].fillna('Unknown', inplace=True)
df['Discount'].fillna('Unknown', inplace=True)
df['Profit'].fillna('Unknown', inplace=True)

```

```python
# Convert date columns to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
sales_by_region = df.groupby('Region')['Sales'].sum()
print(sales_by_region)
```

```python
df.to_csv("Cleaned_Superstore_Sales.csv", index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Correlation heatmap
correlation = df.corr()
sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
